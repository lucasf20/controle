# RESPOSTAS DE SISTEMAS AO DEGRAU – INFLUÊNCIA DE ZEROS
    
Objetivos: Estudar a influência dos zeros (raízes do numerador da função de transferência) na resposta no tempo

## ANALISE DA RESPOSTA NO TEMPO EM PRESENÇA DE ZEROS
            
### Experimento 1
        
Analise as respostas ao degrau em malha aberta, devido a um degrau unitário, dos seguintes sistemas abaixo:

```python
from control import * # pip install control
import matplotlib.pyplot as plt # pip install matplotlib
import numpy

def step_resp(funcao_transferencia): #calcular a resposta, plotar o gráfico e encontrar polos e zeros
    t,y = step_response(funcao_transferencia, T = 50)
    print("Zeros:")
    zeros = zero(funcao_transferencia)
    if(len(zeros)):
        for z in zeros:
            print(z)
    else:
        print("Não há zeros")
    print("\nPolos:")
    polos = pole(funcao_transferencia)
    if(len(polos)):
        for p in polos:
            print(p)
    else:
        print("Não há polos")
        
    max_y = max(y)
    min_y = min(y)
    fig = plt.figure(num=None, figsize=(10,7), dpi=300, facecolor='w', edgecolor='k')
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 52, 2))
    ax.set_yticks(np.append(numpy.arange(min_y, max_y + (max_y-min_y)/20, (max_y-min_y)/20),1))
    plt.grid()
    plt.plot(t,y)
```

a) $\frac{1}{(1 + s)(1 + 5s)} = \frac{1}{5s^2  + 6s + 1}$

```python
item_a = tf(1, [5,6,1])
print(item_a)
```

```python
step_resp(item_a)
```
