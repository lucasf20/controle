2) Para o sistema representado pelo diagrama de blocos na figura abaixo determine os valores de K para os quais há estabilidade.


```python
def checa_k(k):
    func = tf([k],[1,8,9,0])
    print("Funcao:")
    print(func)
    f = feedback(func)
    print("Feedback:")
    print(f)
    ty,tx = step(f)
    plt.plot(tx,ty)
    plt.show()
    
checa_k(0.1)
checa_k(71.9) 
checa_k(0)
checa_k(72)
```

## INFLUÊNCIA DO GANHO
### Experimento 2
Seja o processo

Encontrar os Diagramas de Bode para o processo com os ganhos 5, 20 e 40. Observar que o diagramas de fase são os mesmos. Observar que o diagramas de ganho têm a mesma forma, apenas havendo translação produzida pela variação de ganho. Medir as frequências ωϕ

```python
def bode_ganho(ganho):
    func = tf([1],[1,1.2,1])
    print(ganho*func)
    
    func_fed = feedback(ganho*func)
    print(func_fed)
    
    ty,tx = step(func_fed)
    plt.plot(tx,ty)
    plt.show()
    
    return ganho*func

funcs = [bode_ganho(f) for f in [5,20,40]]
```

```python
for f in funcs:
    print(f)
    fig = plt.figure(num=None, figsize=(10,7), dpi=300, facecolor='w', edgecolor='k')
    bode_plot(f, margins=True, figure=fig)
```

## INFLUÊNCIA DE POLOS E ZEROS
### Experimento 3
a) Encontrar os Diagramas de Bode de

Obs: Observar que a fase está sempre acima de -180, o que implica que para qualquer ganho sempre se tem margem de fase positiva; e, portanto, estabilidade em malha fechada.

```python
def bode_plt(func):
    print(func)
    ty,tx = step(feedback(func))
    plt.plot(tx,ty)
    plt.show
    fig = plt.figure(num=None, figsize=(10,7), dpi=300, facecolor='w', edgecolor='k')
    bode_plot(f, margins=True, figure=fig)
    print("")
    
bode_plt(tf([2],[1,1.2,1]))
```

Houve uma melhora na estabilidade.

```python
bode_plt(2*tf([1,3],[1,1.2,1]))
```
