# RESPOSTAS DE SISTEMAS AO DEGRAU – INFLUÊNCIA DE ZEROS
    
Objetivos: Estudar a influência dos zeros (raízes do numerador da função de transferência) na resposta no tempo

## ANALISE DA RESPOSTA NO TEMPO EM PRESENÇA DE ZEROS
            
### Experimento 1
        
Analise as respostas ao degrau em malha aberta, devido a um degrau unitário, dos seguintes sistemas abaixo:

```python
from control.matlab import * #pip install control
import matplotlib.pyplot as plt # pip install matplotlib
import numpy

def step_resp(funcao_transferencia): #calcular a resposta, plotar o gráfico e encontrar polos e zeros
    y,t = step(funcao_transferencia, T = 50)
    print("Zeros:")
    zeros = zero(funcao_transferencia)
    if(len(zeros)):
        for z in zeros:
            print(z)
    else:
        print("Não há zeros")
    print("\nPolos:")
    polos = pole(funcao_transferencia)
    if(len(polos)):
        for p in polos:
            print(p)
    else:
        print("Não há polos")
        
    max_y = max(y)
    min_y = min(y)
    fig = plt.figure(num=None, figsize=(10,7), dpi=300, facecolor='w', edgecolor='k')
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 52, 2))
    ax.set_yticks(numpy.append(numpy.arange(min_y, max_y + (max_y-min_y)/20, (max_y-min_y)/20),1))
    plt.grid()
    plt.plot(t,y)
    
def parametros_de_desempenho(funcao_transferencia):
    params = stepinfo(funcao_transferencia)
    key_parse = {'RiseTime': 'Rise Time',
                 'SettlingTime': 'Settling Time',
                 'SettlingMin': 'Settling Min',
                 'SettlingMax': 'Settling Max',
                 'Overshoot': 'Overshoot',
                 'Undershoot': 'Undershoot',
                 'Peak': 'Peak',
                 'PeakTime': 'Peak Time',
                 'SteadyStateValue': 'Steady State Value'
                }
    for key in params.keys():
        print(key_parse[key] + ": " + str(params[key]))
```

a) $\frac{1}{(1 + s)(1 + 5s)} = \frac{1}{5s^2  + 6s + 1}$

```python
item_a = tf(1, [5,6,1])
print(item_a)
```

```python
step_resp(item_a)
```

j) Explique as diferenças para cada caso.

Os resultados obtidos nos itens anteriores foram separados em dois clusters de acordo com a sua aparência.

```python
p = []
fig = plt.figure(num=None, figsize=(10,7), dpi=300, facecolor='w', edgecolor='k')
y,t = step(item_a, T = 50)
p.append(plt.plot(t,y,label="item a")[0])
y,t = step(item_b, T = 50)
p.append(plt.plot(t,y,label="item b")[0])
y,t = step(item_c, T = 50)
p.append(plt.plot(t,y,label="item c")[0])
y,t = step(item_d, T = 50)
p.append(plt.plot(t,y,label="item d")[0])
y,t = step(item_e, T = 50)
p.append(plt.plot(t,y,label="item e")[0])
y,t = step(item_f, T = 50)
p.append(plt.plot(t,y,label="item f")[0])
y,t = step(item_g, T = 50)
p.append(plt.plot(t,y,label="item g")[0])
ax = fig.gca()
ax.set_xticks(numpy.arange(0, 52, 2))
ax.set_yticks(numpy.arange(-2, 15, 1))
plt.grid()
plt.legend(handles=p, title='Itens', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='xx-large')
plt.show()
```

No primeiro cluster, temos os itens a,b,c,d,e,f, e g.
    
A primeira observação que podemos fazer é que os itens a, d, e e f se estabilizam em torno do valor 1. Foi observada uma semelhança no numerador, pois todos estes itens possuem um termo independente de valor 1.
    
A segunda observação que temos a fazer é que os itens b e c se estabilizam em um valor em torno de 0. Foi visto que o numerador destes itens não possuem termos independentes e possuem apenas uma raíz igual a 0.
    
A terceira obsevação é que o item g possui um formato muito parecido com o item d, porém o item g possui um ganho igual a 2 e também possui um polinômio de grau 2 no numerador, o que eleva o seu valor de estabilização para em torno de 3.  

```python
p = []
fig = plt.figure(num=None, figsize=(10,7), dpi=300, facecolor='w', edgecolor='k')
y,t = step(item_h, T = 50)
p.append(plt.plot(t,y,label="item h")[0])
y,t = step(item_i, T = 50)
p.append(plt.plot(t,y,label="item i")[0])
ax = fig.gca()
ax.set_xticks(numpy.arange(0, 52, 2))
ax.set_yticks(numpy.append(numpy.arange(-50, 70 + (120)/20, (120)/20),1))
plt.grid()
plt.legend(handles=p, title='Itens', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='xx-large')
plt.show()
```

O segundo cluster é composto pelos itens h e i.

Eles possuem um formato bem distinto dos demais itens, com grandes oscilações no inicio até uma estabilização em torno do valor 1.

Esta estabilização em torno de 1 pode ser explicada pelo mesmo motivo dos itens a,d,e e f do cluster anterior. Ambos os itens deste cluster possuem um termo independente de valor 1 no numerador.

O item i possui valores de pico muito maiores que o item h, o que provavelmente se deve ao valor 8s do item i.
